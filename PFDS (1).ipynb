# DIETICIAN PERSONAL MANAGEMENT SYSTEM

```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def add(df):
    df.to_csv('Diet.csv',mode='a',index=False,header=False)
    print("Data added successfully.")

def update1(df,diet1, diet2, position):
    df.loc[position-1,'Diet']=diet2
    df.to_csv('Diet.csv',mode='a',index=False,header=False)
def update2(df,age1,age2, position):
    df.loc[position-1,'Age']=age2
    df.to_csv('Diet.csv',mode='a',index=False,header=False)
    
def delete(df,num,position):
    if num in df.values:
        df.drop(position-1, axis=0, inplace=True)
        df.to_csv('Diet.csv',mode='a',index=False,header=False)
    else:
        print("No record found with this phone number.")
        
def view(df):
    if(df.empty==True):
        print("No records yet.")
    else:
        print(df[['Name','Phone No.','Age','Gender','Diet']].to_string())
        
def visual(df):
    print("Using Matplotib")
    plt.figure(figsize=(10,10))
    plt.subplot(131)
    plt.bar(df['Age'], df['Gender'])
    plt.title("Bar Chart")
    plt.ylabel("Y axis")
    plt.xlabel("X axis")
    plt.subplot(132)
    plt.scatter(df['Age'], df['Gender'])
    plt.title("Scatter Plot")
    plt.ylabel("Y axis")
    plt.xlabel("X axis")
    plt.subplot(133)
    plt.plot(df['Age'])
    plt.title("Line Graph")
    plt.ylabel("Y axis")
    plt.xlabel("X axis")
    plt.suptitle('Diet Records')
    plt.show() 


        
def main():
    global df1;
    print("\t\t\t\t\t\t\t♦♦♦♦♦♦♦ WELCOME ♦♦♦♦♦♦♦\n")
    print("""CHOOSE WHAT YOU WANT TO DO:-\n1. Add records.\n2. Update records.\n3. Delete records.\n4. View records.\n5. Visualization.\n6. Exit\n""")
    while (True):
        usr_response = int(input("Enter your choice: "))
        if usr_response == 1:
            a1=input("Enter the name:")
            a2=int(input("Enter the phone number:"))
            a3=int(input("Enter the age:"))
            a4=input("Enter the gender [F/M/O]:")
            a5=input("Enter the diet (using commas):")
            a5.split(',')
            df1=pd.DataFrame({'Name':[a1],
                              'Phone No.':[a2],
                              'Age':[a3],
                              'Gender':[a4],
                              'Diet':[a5]})
            ch=input('Do you want to add more? [y/n]: ')
            if ch=='y':
                n=int(input("Enter the number of records you want to enter: "))
                for i in range(n):
                    e1=input("Enter the name:")
                    e2=int(input("Enter the phone number:"))
                    e3=int(input("Enter the age:"))
                    e4=input("Enter the gender [F/M/O]:")
                    e5=input("Enter the diet (using commas):")
                    e5.split(',')
                    df3=pd.DataFrame({'Name':[e1],
                                      'Phone No.':[e2],
                                      'Age':[e3],
                                      'Gender':[e4],
                                      'Diet':[e5]})
                    df1=pd.concat([df1,df3])
                    df1=df1.reset_index()
                add(df1);
            else:
                add(df1);
        elif usr_response == 2:
            res=int(input("What do you want to update: \n1. Diet \n2. Age \n"))
            if res==1:
                a61=input("Enter the previous diet (using commas):")
                a6=input("Enter the updated diet (using commas):")
                a62=int(input('Enter the position'))
                a61.split(',')
                a6.split(',')
                update1(df1,a61,a6,a62);
            elif res==2:
                a7=int(input("Enter the previous age:"))
                a81=int(input("Enter the updated age:"))
                a82=int(input('Enter the position'))
                update2(df1,a7,a81,a82);
            else:
                print("Invalid Choice.")
        elif usr_response == 3:
            a8=int(input("Enter the phone number of the person who's record you want to delete:"))
            a9=int(input('Enter the position'))
            delete(df1,a8,a9);
        elif usr_response == 4:
            view(df1);
        elif usr_response == 5:
            visual(df1);
        elif usr_response == 6:
            print("THANK YOU ! \n")
            exit()
        else:
            print("Invalid Choice.")

main()
```
